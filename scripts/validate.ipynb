```python
from rdkit import Chem
```

SMILES が有効なものか判断し、有効な化合物の数を出力するとともに有効なものだけを別のファイルにまとめます

```python
def is_valid_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return mol is not None

def validate_and_export_smiles(input_file, output_file):
    valid_smiles = []
    invalid_smiles = []

    with open(input_file, 'r') as infile:
        for line in infile:
            smiles = line.strip()
            if is_valid_smiles(smiles):
                valid_smiles.append(smiles)
            else:
                invalid_smiles.append(smiles)
            
    with open(output_file, 'w') as outfile:
        for smiles in valid_smiles:
            outfile.write(smiles + '\n')

    return len(valid_smiles)
```

```python
input_file = '../gen/gen_all.smi'
output_file = '../gen/gen_valid.smi'
val_smiles_num = validate_and_export_smiles(input_file, output_file)

print(f"Valid smiles num: {val_smiles_num}")
```
